# Machine Learning Model Construction

Initially we are not going to look at the spatial component, we will look at time only.

```python
print(hindcast_gc_loc.index[0])
print(hindcast_gc_loc.index[-1])
```

To create a training dataset we will step backwards in time using pandas shift function. For every hs in the dataset we will need to provide training data so it will look something like this:

|t0 |  t-1 |  t-2|   t-3|
|---|---|---|--|
|1.3|  1.2|   1.2 |  1.1|

Where t0 is the actual value and t-1 to-3 are the proceeding records.

We will use the pandas shift function to create this dataset


```python
#create new hs dataframe
hindcast_gc = pd.DataFrame()
hindcast_gc['hs'] = hindcast_gc_loc.hs
```

```python
hindcast_gc['hs-1'] = hindcast_gc.hs.shift(-1)
hindcast_gc['hs-2'] = hindcast_gc.hs.shift(-2)
hindcast_gc['hs-3'] = hindcast_gc.hs.shift(-3)
```

```python
hindcast_gc = hindcast_gc.dropna()
```

```python
hindcast_gc[-5:-1]
```

```python
#input data (before hs)
X = hindcast_gc[[col for col in hindcast_gc.columns if col != 'hs']]

#output data (observed hs)
y = hindcast_gc['hs']  
```

```python
hindcast_gc.info()  
```

```python
from sklearn.model_selection import train_test_split  
```

```python
# split data into training set and a temporary set using sklearn.model_selection.traing_test_split
X_train, X_tmp, y_train, y_tmp = train_test_split(X, y, test_size=0.2, random_state=23)  
```

```python
# take the remaining 20% of data in X_tmp, y_tmp and split them evenly
X_test, X_val, y_test, y_val = train_test_split(X_tmp, y_tmp, test_size=0.5, random_state=23)

X_train.shape, X_test.shape, X_val.shape  
print("Training instances   {}, Training features   {}".format(X_train.shape[0], X_train.shape[1]))  
print("Validation instances {}, Validation features {}".format(X_val.shape[0], X_val.shape[1]))  
print("Testing instances    {}, Testing features    {}".format(X_test.shape[0], X_test.shape[1]))  
```

## Deep Neural Network

we need to create the feature for our machine learning model, this is straight forward using the tensorflow module

```python
feature_cols = [tf.feature_column.numeric_column(col) for col in X.columns]  
```

```python
regressor = tf.estimator.DNNRegressor(feature_columns=feature_cols,  
                                      hidden_units=[50, 50],
                                      model_dir='tf_wx_model')
```

```python
def wx_input_fn(X, y=None, num_epochs=None, shuffle=True, batch_size=400):  
    return tf.estimator.inputs.pandas_input_fn(x=X,
                                               y=y,
                                               num_epochs=num_epochs,
                                               shuffle=shuffle,
                                               batch_size=batch_size)
```

```python
evaluations = []  
STEPS = 400  
for i in range(100):  
    regressor.train(input_fn=wx_input_fn(X_train.reset_index(drop=True), y=y_train.reset_index(drop=True)), steps=STEPS)
    evaluations.append(regressor.evaluate(input_fn=wx_input_fn(X_val.reset_index(drop=True),
                                                               y_val.reset_index(drop=True),
                                                               num_epochs=1,
                                                               shuffle=False)))
```

```python
import matplotlib.pyplot as plt  
%matplotlib inline

# manually set the parameters of the figure to and appropriate size
plt.rcParams['figure.figsize'] = [14, 10]

loss_values = [ev['loss'] for ev in evaluations]  
training_steps = [ev['global_step'] for ev in evaluations]

plt.scatter(x=training_steps, y=loss_values)  
plt.xlabel('Training steps (Epochs = steps / 2)')  
plt.ylabel('Loss (SSE)')  
plt.show()  
```
