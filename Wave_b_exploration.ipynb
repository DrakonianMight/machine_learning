# Exploration of the available wave buoy data for the gold coast site

**Purpose:** an initial look at the gold coast wave buoy data that is available, with the aim of finding some natural limiting factors for machine learning models.

**Date:** 29-12-2018

```python
import pandas as pd
import numpy as np
import sklearn
import os
import matplotlib.pyplot as plt
```

```python
gc_data_folder = r'D:/Masters/data/wave_buoy/gold_coast'
```

```python
wave_b_list = []
for file in os.listdir(gc_data_folder):
    if '.csv' in file:
        filepath = gc_data_folder + "/" + file
        df = pd.read_csv(filepath)
        wave_b_list.append(df)
```

```python
gc_wb = pd.concat(wave_b_list, sort = False)
```

```python
gc_wb.index = pd.to_datetime(gc_wb['Date/Time'])
```

```python
fig, ax = plt.subplots(figsize = (14, 6))
ax.plot(gc_wb.index, gc_wb.Hs)
plt.show()
fig, ax = plt.subplots(figsize = (14, 6))
ax.plot(gc_wb.index, gc_wb.Tz)
plt.show()
```

**comment:** clearly a different method of putting NAN values in the data has been applied in recent years

Remove -99999 values that are errors

```python
gc_wb = gc_wb[gc_wb.Hs > 0]
```

```python
gc_wb.head()
```

```python
gc_wb_plt = gc_wb[['Hs', 'Hmax','Tz','Tp']]
```

```python
gc_wb_plt = gc_wb_plt[gc_wb_plt.Hmax > 0.0]
gc_wb_plt = gc_wb_plt[gc_wb_plt.Tz > 0.0]
gc_wb_plt = gc_wb_plt[gc_wb_plt.Tp > 0.0]
```

Have a look at the underlying data disributions and the relationships between the various wave statistics using a scatter matrix

```python
from pandas import plotting
plotting.scatter_matrix(gc_wb_plt, alpha = 0.2, figsize = (14, 6))
```

Now lets look at wave direction

Datasets need to be merged between older directional data and modern directional data.

```python
gc_wb_dir = gc_wb[['Dir_Tp TRUE', 'Peak Direction']]
```

```python
mod_dir = gc_wb_dir['Peak Direction'].dropna()
```

```python
mod_dir = mod_dir[mod_dir > 0.0]
```

```python
old_dir = gc_wb_dir['Dir_Tp TRUE'].dropna()
```

```python
gc_wb_dir = pd.concat([old_dir, mod_dir])
```

```python
gc_wb_dir = gc_wb_dir[gc_wb_dir > 0.0]
```

```python
gc_wb_plt['pkdir'] = gc_wb_dir
```

Add the cleaned direction data back into the underlying dataset and then plot in scatter matrix for comparison

```python
from pandas import plotting
dir_scatter = plotting.scatter_matrix(gc_wb_plt, alpha = 0.2, figsize = (14, 14))
plt.savefig('gold_coast_matrix.png')
```
